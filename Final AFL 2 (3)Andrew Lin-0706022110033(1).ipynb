# Feature Engineering (Create/Remove feature from available data)

```python
# 1. Age Extraction
current_year = pd.to_datetime('today').year
df['Age'] = current_year - df['Year_Birth']
```

```python
# 2. Spending Power
categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Spent'] = df[categories].sum(axis=1)
```

```python
# 3. Household Living Arrangement
df['Living_With'] = df['Marital_Status'].replace({
    'Married': 'With Spouse',
    'Together': 'With Partner',
    'Single': 'Single',
    'Divorced': 'Divorced',
    'Widow': 'Widow'
})
```

```python
# 4. Family Dynamics
df['Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# 5. Family Unity
df['Family_Size'] = df['Children'] + 2  # Dua (orang tua) ditambah jumlah anak
```

```python
# 6. Parental Status
df['Is_Parent'] = df['Children'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# 7. Educational Evolution
df['Education'] = df['Education'].replace({
    'Basic': 'Basic Education',
    '2n Cycle': '2nd Cycle Education',
    'Graduation': 'Graduation',
    'Master': 'Master',
    'PhD': 'PhD'
})
```

```python
# 8. Farewell to Redundancy
redundant_features = ['ID', 'Year_Birth', 'Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Marital_Status'] 
df.drop(columns=redundant_features, inplace=True)
```

# Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
kmeans_params = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}
SSE = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, **kmeans_params)
    kmeans.fit(df_pca)
    SSE.append(kmeans.inertia_)
plt.plot(k_values, SSE, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.title("Elbow Method for Optimal K")
plt.grid()
plt.show()
```

```python
from kneed import KneeLocator
kneedle = KneeLocator(range(1, 11), SSE, curve="convex", direction="decreasing")
print(f'the best k for your model is {kneedle.elbow}')
```

```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
n_clusters = 3  
model = AgglomerativeClustering(n_clusters=n_clusters)
cluster_labels = model.fit_predict(df_pca)
df_pca['Cluster_Label'] = cluster_labels
print(df_pca)
```
